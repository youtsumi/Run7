```python
from astropy.io import fits
from astropy.stats import mad_std
import pylab
import numpy
import matplotlib.pyplot as plt
```

```python
fh = fits.open('/sdf/data/rubin/offline/ucd/20241001/TS_C_20241001_000002_R21_S01.fits')
```

```python
fig,axs=plt.subplots(2,8,figsize=(12,10))
for hdu, ax in zip(fh[1:], axs.flatten()):
    img = hdu.data
    std=mad_std(img)
    mean=numpy.median(img)
    ax.imshow(img,origin='lower',vmin=mean-std,vmax=mean+std)
    ax.set_title(hdu.header['EXTNAME'])
    ax.axis('off')
fig.suptitle(f"{fh[0].header['OBSID']}")
fig.tight_layout()
```
